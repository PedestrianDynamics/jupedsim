# Overview performace stats

```python
import sqlite3
import pandas as pd

# point this to your results sqlite to analyze
input = 'large_street_network.sqlite'

cnx = sqlite3.connect(input)
df = pd.read_sql_query("SELECT * FROM perf_statistics", cnx)
cnx.close()
```

## Total iteration time / agent count per iteration

```python
df['iteration_loop_us'].plot(figsize=(20,10), xlabel="iteration", ylabel="time[µs]", legend=True)
df['agent_count'].plot(secondary_y=True, ylabel="agents", legend=True)
```

## Time to compute oerational level update / agent count per iteration

```python
df['operational_level_us'].plot(figsize=(20,10), xlabel="iteration", ylabel="time[µs]", legend=True)
df['agent_count'].plot(secondary_y=True, ylabel="agents", legend=True)
```

## Total time w.o. operational level / agent count per iteration

```python
df['delta'] = df['iteration_loop_us'] - df['operational_level_us']
df['delta'].plot(figsize=(20,10), xlabel="iteration", ylabel="time[µs]", legend=True)
df['agent_count'].plot(secondary_y=True, ylabel="agents", legend=True)
```
